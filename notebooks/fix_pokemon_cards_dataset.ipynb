## Description
The [TheFusion21/PokemonCards](https://huggingface.co/datasets/TheFusion21/PokemonCards) dataset on HuggingFace Hub has some broken image URLs. This notebook iterates through the dataset examples, discards the invalid examples, and uploads the dataset on the Hub.

```python
from huggingface_hub import notebook_login

notebook_login()
```

```python
import requests
from tqdm.auto import tqdm
from datasets import load_dataset
```

```python
ds_path = "TheFusion21/PokemonCards"
```

```python
ds = load_dataset(ds_path, split="train")
```

```python
# Example of broken image URL:
requests.get("https://images.pokemontcg.io/mcd17/2_hires.png", stream=True)
```

```python
valid_indices = []

for idx, x in tqdm(enumerate(ds), total=ds.num_rows):
    response = requests.get(x["image_url"], stream=True)
    try:
        response.raise_for_status()
        valid_indices.append(idx)
    except:
        print(f"Invalid example {idx}")
        continue
```

```python
ds_filtered = ds.select(valid_indices)
```

```python
ds_filtered.push_to_hub("tonywu71/PokemonCards_fixed", split="all", private=True)
```
